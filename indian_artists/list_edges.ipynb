```python
import wikipedia
import csv
import json
```

```python
pages = set()

with open('nodes.csv', newline='') as nodes_csv_file:
    reader = csv.DictReader(nodes_csv_file)
    for row in reader:
        pages.add(row['node'])

```

```python
pages_visited_outside_category = set()
pages_visited_outside_category_with_more_hops = set()
pages_visited_inside_category = set()
# graph = {}
# count = 0
graph_csv_fields = ["node", "hop_from_category", "edges_within_category", "edges_outside_category"]

for page in pages:
    if page not in pages_visited_inside_category:
        pages_visited_inside_category.add(page)
        links = []
        out_links = []
        try:
            for link in wikipedia.WikipediaPage(title=page).links:
                # print(link)
                if link in pages:
                    links.append(link)
                elif link in pages_visited_outside_category:
                    out_links.append(link)
                elif link not in pages_visited_outside_category_with_more_hops:
                    try:
                        consider = False
                        for next_link in wikipedia.WikipediaPage(title=link).links:
                            if next_link in pages:
                                consider = True
                                with open("nodes_outside.csv", 'a', newline='') as nodes_outside_csv_file:
                                    writer = csv.DictWriter(nodes_outside_csv_file, fieldnames=["node"])
                                    writer.writerow({'node': link})
                                pages_visited_outside_category.add(link)
                                out_links.append(link)
                                break
                        if consider == False:
                            pages_visited_outside_category_with_more_hops.add(link)
                    except Exception as e:
                        print(e)

            # if count % 100 == 0:
            #     logging.info(str(count)+"/"+str(len(pages))+". Edges successfully added for page: "+ page)
        except Exception as e:
            print(e)
            # logging.error(str(count)+"/"+str(len(pages))+". Edges failed for page: "+ page + str(e))
        with open("graph.csv", 'a', newline='') as graph_csv_file:
            writer = csv.DictWriter(graph_csv_file, fieldnames=graph_csv_fields)
            writer.writerow({'node': page,
                            'hop_from_category': 0,
                            'edges_within_category': json.dumps({"key":links}),
                            'edges_outside_category': json.dumps({"key":out_links})
            })
        # print(page)
```
